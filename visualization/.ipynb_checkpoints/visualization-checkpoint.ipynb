```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the merged dataset
merged_df = pd.read_csv('../Information Mart/orders_products_details.csv')

# Step 2: Explore the data (optional)

# Step 3: Create visualizations


# Visualization 2: Top Selling Products
top_products = merged_df.groupby('product_name')['quantity'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar', color='skyblue')
plt.title('Top Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.savefig('top_selling_products.png')
plt.show()

# Visualization 3: Revenue by Product Category
plt.figure(figsize=(10, 6))
sns.barplot(data=merged_df, x='category_id', y='list_price_x', estimator=sum, ci=None)
plt.title('Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.savefig('revenue_by_product_category.png')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
merged_df[merged_df['list_price_egp'] < 150000]['list_price_egp'].plot.hist()
plt.title('Distribution of Sold Product Prices')
plt.xlabel('Product Price (EGP)')
plt.ylabel('Frequency')
plt.savefig('product_price_distribution.png')
plt.show()
```
