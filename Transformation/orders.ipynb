```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv('../staging_1/orders/orders.csv')
orders
```

```python
orders['late_delivery'] = orders['shipped_date'] > orders['required_date']
orders
```

```python
orders['required_date'] = pd.to_datetime(orders['required_date'], format='%Y-%m-%d', errors='coerce')
orders['shipped_date'] = pd.to_datetime(orders['shipped_date'], format='%Y-%m-%d', errors='coerce')
orders['latency_days'] = (orders['shipped_date'] - orders['required_date']).dt.days
orders.loc[orders['latency_days'] < 0, 'latency_days'] = np.nan
orders
```

```python
orders['day_of_week'] = orders['shipped_date'].dt.dayofweek
orders['weekend'] = (orders['day_of_week'] == 6) | (orders['day_of_week'] == 0)
orders.drop(columns=['day_of_week'], inplace=True)
orders
```

```python
orders.to_csv('../staging_2/orders/orders.csv', index = False)
```
