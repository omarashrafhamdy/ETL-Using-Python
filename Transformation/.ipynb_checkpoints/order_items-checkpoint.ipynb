```python
import pandas as pd
```

```python
exchange = pd.read_csv('../staging_1/exchange/exchange.csv')
exchange
```

```python
order_items = pd.read_csv('../staging_1/order_items/order_items.csv')
order_items
```

```python
order_items_exchange = pd.merge(order_items,exchange,how='cross')
order_items_exchange['list_price_egp'] = order_items_exchange['list_price'] * order_items_exchange['EGP']
order_items_exchange
```

```python
order_items_exchange = order_items_exchange.drop(['timestamp_extraction', 'source_y', 'timestamp'], axis=1)
order_items_exchange
```

```python
order_items_exchange.to_csv('../staging_2/order_items/order_items.csv', index = False)
```
