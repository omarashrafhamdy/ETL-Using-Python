```python
import pandas as pd
```

```python
orders = pd.read_csv('../staging_2/orders/orders.csv')
orders
```

```python
products = pd.read_csv('../staging_2/products/products.csv')
products
```

```python
order_items = pd.read_csv('../staging_2/order_items/order_items.csv')
order_items
```

```python
orders_items = pd.merge(order_items , orders , on='order_id' , how='inner')
orders_items
```

```python
orders_products_details = pd.merge(orders_items , products , on='product_id' , how='inner')
orders_products_details
```

```python
orders_products_details.to_csv('../Information Mart/orders_products_details.csv', index = False)
```
